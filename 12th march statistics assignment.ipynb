# Q1

```python
import scipy.stats as stats
sample_mean = 50
sample_stddev = 5
sample_size = 20
confidence_level = 0.95
degrees_of_freedom = sample_size - 1
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df=degrees_of_freedom) * (sample_stddev / (sample_size ** 0.5))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print("95% Confidence Interval:", confidence_interval)
```

# Q2

```python
import scipy.stats as stats
import numpy as np
observed_frequencies = [100,40,20,130,50,20]  
expected_probabilities = [0.20, 0.20, 0.20, 0.10, 0.10, 0.20]
total_observations = sum(observed_frequencies)
expected_frequencies = [expected_prob * total_observations for expected_prob in expected_probabilities]
chi_square_statistic, p_value = stats.chisquare(observed_frequencies, expected_frequencies)
alpha = 0.05
print("Chi-square statistic:", chi_square_statistic)
print("P-value:", p_value)
if p_value < alpha:
    print("Reject the null hypothesis. The distribution of colors does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The distribution of colors matches the expected distribution.")
```

# Q3

```python
import scipy.stats as stats
import numpy as np
observed_data = np.array([[20, 15], [10, 25], [15, 20]])
chi_square_statistic, p_value, dof, expected = stats.chi2_contingency(observed_data)
print("Chi-square statistic:", chi_square_statistic)
print("P-value:", p_value)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant association between the groups and outcomes.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between the groups and outcomes.")
```

# Q4

```python
import scipy.stats as stats
import numpy as np
sample_size = 500
observed_smokers = 60
sample_proportion = observed_smokers / sample_size
z = stats.norm.ppf(0.975)  
margin_of_error = z * np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)
confidence_interval = (sample_proportion - margin_of_error, sample_proportion + margin_of_error)
print("95% Confidence Interval:", confidence_interval)
```

# Q6

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
df = 10
x = np.linspace(0, 30, 1000)
pdf = stats.chi2.pdf(x, df)
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f"Chi-square Distribution (df={df})")
plt.fill_between(x, pdf, where=(x >= 0) & (x <= 15), color='gray', alpha=0.5, label='Area for chi-square statistic of 15')
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density')
plt.title(f'Chi-square Distribution with {df} Degrees of Freedom')
plt.legend()
plt.grid(True)
plt.show()
```

# Q7

```python
import scipy.stats as stats
import numpy as np
sample_size = 1000
observed_coke_preferred = 520
sample_proportion = observed_coke_preferred / sample_size
z = stats.norm.ppf(0.995)  # 0.995 to account for two-tailed distribution
margin_of_error = z * np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size + (z**2 / (4 * sample_size**2)))
confidence_interval = (sample_proportion - margin_of_error, sample_proportion + margin_of_error)
print("99% Confidence Interval:", confidence_interval)
```

# Q8

```python
import scipy.stats as stats
import numpy as np
observed_tails = 45
total_trials = 100
expected_prob_tails = 0.5
expected_tails = expected_prob_tails * total_trials
expected_heads = total_trials - expected_tails
observed_frequencies = np.array([observed_tails, total_trials - observed_tails])
expected_frequencies = np.array([expected_tails, expected_heads])
chi_square_statistic, p_value = stats.chisquare(observed_frequencies, expected_frequencies)
alpha = 0.05
print("Chi-square statistic:", chi_square_statistic)
print("P-value:", p_value)
if p_value < alpha:
    print("Reject the null hypothesis. The coin may be biased towards tails.")
else:
    print("Fail to reject the null hypothesis. The coin does not appear to be biased towards tails.")
```

# Q11

```python
import scipy.stats as stats
sample_mean = 72
sample_stddev = 10
sample_size = 30
population_mean = 70
significance_level = 0.05
t_statistic = (sample_mean - population_mean) / (sample_stddev / (sample_size ** 0.5))
degrees_of_freedom = sample_size - 1
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=degrees_of_freedom))
print("Calculated t-statistic:", t_statistic)
print("Degrees of freedom:", degrees_of_freedom)
print("P-value:", p_value)
if p_value < significance_level:
    print("Reject the null hypothesis. The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the population mean.")
```
