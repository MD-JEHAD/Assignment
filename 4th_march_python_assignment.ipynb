# Q1

```python
import seaborn as sns
import plotly.graph_objects as go
df = sns.load_dataset("titanic")
fig = go.Figure()
fig.add_trace(go.Scatter(x = df.age,y = df.fare,mode = "markers"))
fig.show()
```

# Q2

```python
import plotly.express as px
df = sns.load_dataset("tips")
fig = px.box(df,x = df["tip"],y=df["size"])
fig.show()
```

## Q3

```python
import plotly.express as px
df = sns.load_dataset("tips")
fig = px.histogram(df,x = df["sex"],y=df["total_bill"],color=df["day"],pattern_shape=df["smoker"])
fig.show()
```

# Q5

Distplot is a function that provides access to several approached for visualizing univariate and bivariate distribution of data.

```python
import plotly.figure_factory as ff
import numpy as np
a = np.random.randn(200)
b = np.random.randn(200)
c = np.random.randn(200)
hist_data = [a, b, c]
group_labels = ['Group 1', 'Group 2', 'Group 3']
fig = ff.create_distplot(hist_data, group_labels,)
fig.show()
```
